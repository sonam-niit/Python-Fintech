```python
import pandas as pd
data= {'country':['Belgium','France','Germany','Netherlands','UK'],
      'population':[11.3,64.3,81.3,16.9,64.9],
      'area':[30510,671308,357050,41256,244820],
      'capital':['Brussels','Paris','Berlin','Amsterdam','London']}
countries=pd.DataFrame(data)
countries.head()
```

```python
data2={'country':['Nigeria','Rwanda','Egypt','Morocco'],
      'population':[182.2,11.3,94.3,34.4],
      'area':[30510,671308,357050,41256],
      'capital':['Abuja','Kingali','Cario','Rabat']}
countries_africa=pd.DataFrame(data2)
countries_africa.head()
```

```python
pd.concat([countries,countries_africa])
```

```python
pd.concat([countries,countries_africa],ignore_index=True)
```

```python
#some datasets which don't have same set of columns, by default missing values get introduced
pd.concat([countries,countries_africa[['country','capital']]],ignore_index=True)
```

```python
#combine the sets by passing dictionary object
pd.concat({'Europe':countries,'Africa':countries_africa})
```
