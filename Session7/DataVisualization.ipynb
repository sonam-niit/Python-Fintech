```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

iris= pd.read_csv('files/iris.csv',names=['sepal_length','sepal_width','petal_length','petal_width','class'])
iris.head()
```

```python
iris_updated=iris.drop('Id')
iris_updated.head()
```

```python
#scatter Plot
fig, ax= plt.subplots()

ax.scatter(iris_updated['sepal_length'],iris_updated['sepal_width'])
#set titles and labels
ax.set_title('Iris DataSet')
ax.set_xlabel('Sepal_length')
ax.set_ylabel('Sepal_width')

plt.show()
```

```python
#create color 
colors={'Iris-setosa':'r','Iris-versicolor':'g','Iris-virginica':'b'}
fig,ax=plt.subplots()

for i in range(len(iris_updated['sepal_length'])):
    ax.scatter(iris_updated['sepal_length'][i],iris_updated['sepal_width'][i],color=colors[iris_updated['class'][i]])

ax.set_title('Iris DataSet')
ax.set_xlabel('Sepal_length')
ax.set_ylabel('Sepal_width')

plt.show();
```

```python
#Line Chart
iris= pd.read_csv('files/iris.csv')
columns= iris.columns.drop(['Id','Species'])
x_data=range(0,iris.shape[0])
fig,ax=plt.subplots()

for c in columns:
    ax.plot(x_data,iris[c])
ax.set_title("Iris DataSet")
ax.legend(labels=['SL','SW','PL','PW'])

plt.show();
```

```python
import yfinance as yf

#set the start date and end date
start_date='1990-01-01'
end_date='2022-12-31'

#set ticker
ticker='AMZN'

#get Data
data= yf.download(ticker,start_date,end_date)

data.tail()
```

```python
data['Adj Close'].plot()
plt.show()
```

```python
data['Adj Close'].plot(figsize=(10,7))

plt.title("Adjusted Close Price %s" % ticker,fontsize=14)

plt.ylabel('Price')
plt.xlabel('Year')

plt.grid(which="major",color='k',linestyle='-.',linewidth=0.4)

plt.show()
```
