```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

iris= pd.read_csv('files/iris.csv',names=['sepal_length','sepal_width','petal_length','petal_width','class'])
iris.head()
```

```python
iris_updated=iris.drop('Id')
iris_updated.head()
```

```python
#scatter Plot
fig, ax= plt.subplots()

ax.scatter(iris_updated['sepal_length'],iris_updated['sepal_width'])
#set titles and labels
ax.set_title('Iris DataSet')
ax.set_xlabel('Sepal_length')
ax.set_ylabel('Sepal_width')

plt.show()
```
